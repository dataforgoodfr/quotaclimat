```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/mediatree_channel_coverages_2026-02-02")
df.timestamp = pd.to_datetime(df.timestamp)
```

```python
df.timestamp.max().strftime("%d %b")
```

```python
import plotly.express as px
import os

for country, group in df.groupby("country"):
    start_date = group.timestamp.min().strftime("%d %B")
    end_date = group.timestamp.max().strftime("%d %B")
    fig = px.line(group, x="timestamp", y="coverage", color='channel_name', title=f"{country.title()}: {start_date} - {end_date}")
    os.makedirs(f"images/{group.timestamp.min().strftime('%d%b')}_{group.timestamp.max().strftime('%d%b')}", exist_ok=True)
    fig.write_image(f"images/{group.timestamp.min().strftime('%d%b')}_{group.timestamp.max().strftime('%d%b')}/coverage_{country}_chains_{group.timestamp.min().strftime('%d%b')}_{group.timestamp.max().strftime('%d%b')}_timestamps.png")
    fig.show()
```

```python
import plotly.express as px
import os

df.timestamp = df.timestamp.dt.replace(hour=0, minute=0)
df = df.groupby(["timestamp", "country", "channel_name"]).agg(
    {"coverage": "mean", "total_results": "sum", "total_minutes": "sum"}
)

for country, group in df.groupby("country"):
    start_date = group.date.min().strftime("%d %B")
    end_date = group.date.max().strftime("%d %B")
    fig = px.line(group, x="date", y="coverage", color='channel_name', title=f"{country.title()}: {start_date} - {end_date}")
    os.makedirs(f"images/{group.date.min().strftime('%d%b')}_{group.date.max().strftime('%d%b')}", exist_ok=True)
    fig.write_image(f"images/{group.date.min().strftime('%d%b')}_{group.date.max().strftime('%d%b')}/coverage_{country}_chains_{group.date.min().strftime('%d%b')}_{group.date.max().strftime('%d%b')}.png")
    fig.show()
```

```python
for country, group in df.groupby("country"):
    start_date = group.date.min().strftime("%d %B")
    end_date = group.date.max().strftime("%d %B")
    df_mean = group.groupby("date").agg({"coverage": "mean"})
    fig = px.line(df_mean, y="coverage", title=f"Mean Coverage {country.title()}: {start_date} - {end_date}")
    fig.write_image(f"images/{group.date.min().strftime('%d%b')}_{group.date.max().strftime('%d%b')}/coverage_{country}_mean_{group.date.min().strftime('%d%b')}_{group.date.max().strftime('%d%b')}.png")
    fig.show()
```
