# Comparer deux fichiers

```python
data_climat = read_and_format_one('../data/keywords/20221015_lastmonth_all_changement climatique.xlsx',"../data/channels.xlsx")
data_biodiv = read_and_format_one('../data/keywords/20221015_lastmonth_all_biodiversité.xlsx',"../data/channels.xlsx")

data_agg = pd.concat([data_climat,data_biodiv],axis = 0,ignore_index = True)
```

```python
show_mentions_over_time(data_agg,freq = "D",split = "keyword")
```

```python
show_mentions_over_time(data_agg,freq = "D",split = "keyword",as_percent = True).show()
show_mentions_over_time(data_agg,freq = "D",split = "keyword",kind = "area").show()
```

```python
show_mentions_by_channel(data_agg,list_of_channels = TOP_25_CHANNELS,split = "keyword")
```

```python
show_mentions_treemap(data_agg,TOP_25_CHANNELS,freq = "4H",path = ["channel_name","keyword","time_of_the_day"])
```

```python
show_mentions_treemap(data_agg,TOP_25_CHANNELS,freq = "4H",path = ["media","channel_name","keyword","time_of_the_day"])
```

```python
show_mentions_treemap(data_agg,TOP_25_CHANNELS,freq = "4H",path = ["channel_name","keyword"])
```
