## Schema
- 'ID' : index in table 
- 'Publication Name': publication author
- 'News Title': title of the article
- 'Download Date': date of the ingestion in the database (scrapping date)
- 'News Publication Date': date of the publication of the article
- 'News Keywords': when available, key topic in the article (given by the publication)
- 'Section': article section (e.g., climat, politic)
- 'Image Caption': when available, caption of the image
- 'Media Type': tv or press
- 'URL': url of the article
- 'Updated On': last article update
- 'News Description': Not available yet
