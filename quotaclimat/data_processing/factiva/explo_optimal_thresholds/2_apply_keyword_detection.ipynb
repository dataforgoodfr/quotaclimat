```python
import pandas as pd

from quotaclimat.data_processing.factiva.explo_optimal_thresholds.utils_explo.utils_explo import (
    apply_keyword_detection_to_dataframe,
)

DATA_PATH_TEST2 = "quotaclimat/data_processing/factiva/explo_optimal_thresholds/data_explo/factiva_data_test_llm.xlsx"
DATA_PATH_TEST = 'quotaclimat/data_processing/factiva/explo_optimal_thresholds/data_explo/factiva_data_test.xlsx'
EXPORT_PATH = 'quotaclimat/data_processing/factiva/explo_optimal_thresholds/data_explo/factiva_keywords.xlsx'
```

```python
# Load the data
df = pd.read_excel(DATA_PATH_TEST2, index_col=None).drop(columns=["Unnamed: 0"], errors="ignore")
df_test_original = pd.read_excel(DATA_PATH_TEST, sheet_name='first_test')
df = pd.concat([df_test_original, df])
df = df.drop_duplicates(subset='an', keep='last').reset_index(drop=True)
```

```python
# Apply keyword detection on the 'text' column
# Add all keyword-related columns (counts, lists, aggregations)
df_keywords_assigned = apply_keyword_detection_to_dataframe(df, text_col="text")
```

```python
df_keywords_assigned.to_excel(EXPORT_PATH, sheet_name='keywords')
```
